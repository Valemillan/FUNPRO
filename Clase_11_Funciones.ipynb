#FUNCIONES
EStructura de programación que permiten "cortar" una código grande en partes pequeñas y funcionales.


sintaxis de una función
```
def nombre_de_la_función(d,d1,d2,d2,...,dn):
 operación1
 operación2
 función1
 estructura de control
 estructura de selección
 return variable
```
donde:
* d,d1,d2: son los parámetros de la función
* Por dentro pueden ir todas las operaciones
* return vuelve una variable local --> global


Existen 4 tipos de funciones:
1. Sin parámetro, sin retorno.
```
def función1():
 operación1
 operación2
```
2. Sin parámetro, con retorno.
```
def función2():
 operación1
 operación2
```
3. Con parámetro, sin retorno.
```
def función3(color):
 operación1
 operación2
```
4. Con parámetro, con retorno.
```
def función4(n1,n2,n3,n4):
 operación1
 operación2
 return promedio
```

````python
#Ejemplo: Desarrollar un programa que de n notas (donde n es mayor que 1), me diga cuál es la mayor nota y cuál es la menor
notas=int(input("Ingrese la cantidad de notas: "))
nota=[]
for i in range(notas):
  n=float(input(f"Ingrese la nota {i+1}: "))
  nota.append(n)
print(nota)
````

````python
nota_max=max(nota)
nota_min=min(nota)
print(f"La nota máxima es: {nota_max} y la peor nota es {nota_min}")
````

````python
def fnotas():
  fnotas=int(input("Ingrese la cantidad de notas: "))
  fnota=[]
  for i in range(notas):
   fn=float(input(f"Ingrese la nota {i+1}: "))
   fnota.append(fn)
  return fnota
````

````python
#Función de tipo con parámetro, sin retorno
def max_min(notas):
  nota_max=max(notas)
  nota_min=min(notas)
  print(f"La nota máxima es {nota_max} y la peor nota es {nota_min}")
````

````python
def mano():
  print("manos izquierda en la cabeza")
def piernas(n):
  print("piernas hacia adelante y un poco flexionada")
  for i in range(n):
    print("mover de derecha a izquierda")
def paso1(n,veces):
  for i in range(veces):
    print("La mano izquierda darle la vuelta a la cabeza")
    piernas(n)
    print("subir los antebrazos con las manos flexionados y con los puños cerrados")
    print("abrir y cerrar un poco los puños dos veces")
    print("bajar los brazos")
def paso2(n,veces,p):
  for i in range(p):
    print("subir el brazo izquierdo con el dedo indice hacia arriba y luego el brazo derecho con el dedo indice hacia arriba")
    print("unir los dedos y luego separarlos")
    print("bajar el brazo derecho")
    print("la mano izquierda colocarla de lado al cuerpo y luego bajarla")
    paso1(n,veces)
def baile_mamado(n,veces,p,b):
  mano()
  print("esperar unos segundos antes de iniciar el baile")
  for i in range(b):
    paso1(n,veces)
    paso2(n,veces,p)
def remate():
  print("moverse la cintura de izquierda a derecha")
````

````python
baile_mamado(4,2,2,2)
remate()
````
